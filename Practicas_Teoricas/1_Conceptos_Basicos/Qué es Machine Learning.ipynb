### Lección 1: ¿Qué es Machine Learning?

**Machine Learning** (ML) es una rama de la inteligencia artificial que se enfoca en desarrollar algoritmos que permiten a las máquinas aprender de los datos y hacer predicciones o decisiones sin ser programadas explícitamente para cada tarea.

En términos simples, en lugar de programar reglas para resolver un problema, entrenamos modelos que aprenden a partir de ejemplos (datos).

#### Tipos de Machine Learning
1. **Aprendizaje Supervisado**: El modelo se entrena con datos que ya están etiquetados. Es decir, cada entrada viene con su salida correcta, y el objetivo del modelo es aprender a mapear entradas a salidas. Ejemplos:
   - Clasificación: Predecir si un correo es spam o no.
   - Regresión: Predecir el precio de una casa basada en características como tamaño, ubicación, etc.

2. **Aprendizaje No Supervisado**: El modelo trabaja con datos sin etiquetar. En este caso, busca patrones o relaciones entre los datos por sí mismo. Ejemplos:
   - Agrupamiento (*Clustering*): Agrupar clientes con comportamientos similares.
   - Reducción de dimensionalidad: Simplificar datos manteniendo la mayor cantidad posible de información.

3. **Aprendizaje por Refuerzo**: Un agente aprende a través de la interacción con un entorno. Se premia o penaliza al agente según sus acciones, y su objetivo es maximizar su recompensa a lo largo del tiempo.

### Ejercicio Práctico: Visualización de Datos
Como primer ejercicio, vamos a visualizar un conjunto de datos usando **Pandas** y **Matplotlib**.

**Código:**

Este código carga un dataset clásico de flores (Iris), lo convierte en un DataFrame de Pandas y luego crea un gráfico de dispersión de dos de las características (longitud y ancho del sépalo).


```python
import pandas as pd  # Importa la librería Pandas, que es utilizada para la manipulación y análisis de datos.
import matplotlib.pyplot as plt  # Importa la librería Matplotlib, que se utiliza para generar gráficos.

# Cargar un dataset de ejemplo (en este caso, el dataset 'Iris', que viene incluido en scikit-learn).
from sklearn.datasets import load_iris  # Importa la función 'load_iris' de la librería scikit-learn.
data = load_iris()  # Llama a la función 'load_iris' para cargar el dataset de Iris en la variable 'data'.

# Crear un DataFrame de Pandas a partir de los datos del dataset 'Iris'.
# 'data.data' contiene los datos numéricos de las características (longitud del sépalo, ancho del sépalo, etc.).
# 'data.feature_names' contiene los nombres de las características que se usarán como nombres de columnas.
df = pd.DataFrame(data.data, columns=data.feature_names)  # Crea un DataFrame con los datos y les asigna los nombres de las columnas.

df  # Muestra el DataFrame, que contiene las características del dataset 'Iris'.
```

```python
# Renombrar las columnas al español
df.columns = ['longitud del sépalo (cm)', 'ancho del sépalo (cm)', 'longitud del pétalo (cm)', 'ancho del pétalo (cm)']
df
```

```python
# Visualizar las primeras filas del dataset
df.head()  # Muestra las primeras 5 filas del DataFrame 'df'. Esto es útil para tener una vista rápida de cómo son los datos.

# Crear un gráfico simple de dos características
plt.scatter(df['longitud del sépalo (cm)'], df['ancho del sépalo (cm)'])  
# Crea un gráfico de dispersión (scatter plot) usando las columnas 'longitud del sépalo (cm)' y 'ancho del sépalo (cm)' del DataFrame. 
# Los puntos en el gráfico representarán la relación entre estas dos características.

plt.xlabel('Longitud del Sépalo (cm)')  # Etiqueta el eje X como "Longitud del Sépalo (cm)".
plt.ylabel('Ancho del Sépalo (cm)')  # Etiqueta el eje Y como "Ancho del Sépalo (cm)".
plt.title('Gráfico de dispersión de Longitud vs Ancho del Sépalo')  # Establece el título del gráfico.

plt.show()  # Muestra el gráfico en pantalla. Sin esta función, el gráfico no se visualizará.
```
