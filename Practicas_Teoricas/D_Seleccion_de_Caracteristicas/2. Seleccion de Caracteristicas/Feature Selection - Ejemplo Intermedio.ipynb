**Método de Filtrado**:


```python
#Chi-Cuadrado para Selección de Características

from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Seleccionar las 2 mejores características usando Chi-Cuadrado como métrica de evaluación
selector_filtro = SelectKBest(score_func=chi2, k=2)
X_filtrado = selector_filtro.fit_transform(X, y)

print("Características seleccionadas usando Chi-Cuadrado:")
print(X_filtrado)
```

```python
# Mutual Information para Selección de Características
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, mutual_info_classif

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Seleccionar las 2 mejores características usando Mutual Information como métrica de evaluación
selector_filtro = SelectKBest(score_func=mutual_info_classif, k=2)
X_filtrado = selector_filtro.fit_transform(X, y)

print("Características seleccionadas usando Mutual Information:")
print(X_filtrado)
```

**Método "Wrapper":**

```python
#RFE con SVM para Selección de Características

from sklearn.datasets import load_iris
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Crear un modelo de clasificación SVM
modelo_svm = SVC(kernel="linear")

# Seleccionar las 2 mejores características usando RFE (Recursive Feature Elimination) con SVM
selector_wrapper_svm = RFE(estimator=modelo_svm, n_features_to_select=2)
X_wrapper_svm = selector_wrapper_svm.fit_transform(X, y)

print("Características seleccionadas usando RFE con SVM:")
print(X_wrapper_svm)
```

```python
#RFECV con Regresión Lineal para Selección de Características con Validación Cruzada
from sklearn.datasets import load_iris
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Crear un modelo de regresión lineal
modelo_lineal = LinearRegression()

# Seleccionar las mejores características usando RFECV (Recursive Feature Elimination with Cross-Validation)
selector_rfecv = RFECV(estimator=modelo_lineal, step=1, cv=5)
X_rfecv = selector_rfecv.fit_transform(X, y)

print("Características seleccionadas usando RFECV con Regresión Lineal:")
print(X_rfecv)
```

**Método Embebido:** 

```python
#Selección de Características con Lasso
from sklearn.datasets import load_iris
from sklearn.linear_model import LassoCV

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Crear un modelo de Lasso para selección de características
modelo_lasso = LassoCV()
modelo_lasso.fit(X, y)

# Obtener coeficientes no nulos (características seleccionadas)
coeficientes_lasso = modelo_lasso.coef_

print("Características seleccionadas usando L1 Regularización (Lasso):")
print(coeficientes_lasso)
```

```python
#Selección de Características con Gradient Boosting
from sklearn.datasets import load_iris
from sklearn.ensemble import GradientBoostingClassifier

# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target

# Crear un modelo de clasificación con Gradient Boosting
modelo_gradient_boosting = GradientBoostingClassifier(n_estimators=100, random_state=42)
modelo_gradient_boosting.fit(X, y)

# Obtener importancias de características
importancias_caracteristicas_gb = modelo_gradient_boosting.feature_importances_

print("Importancias de características usando Gradient Boosting:")
print(importancias_caracteristicas_gb)
```
